```python
import pyvisa

rm = pyvisa.ResourceManager()
rm.list_resources()

# Estabelecer ligação com a laser box
laserbox = rm.open_resource('ASRL7::INSTR')

# Definição do baudrate
laserbox.baud_rate = 115200

# Definição dos carateres de fim de mensagem (tanto de "read" como "write")
laserbox.read_termination = "\r\n"
laserbox.write_termination = "\r\n"

laserbox.write("SOURce2:AM:STATe OFF")
laserbox.write("SOURce3:AM:STATe OFF")
laserbox.write("SOURce4:AM:STATe OFF")
laserbox.write("SOURce5:AM:STATe OFF")
```

```python
# INITIALIZATION

import ctypes
import os
import numpy as np

# Load the board dll
dll_path = "RTC5DLLx64"

board = ctypes.windll.LoadLibrary(dll_path)

# Initializing the dll
init_result = board.init_rtc5_dll()
print("Initialization:", init_result)

# Mode
mode = board.set_rtc4_mode()
print("Mode:", mode)

execution = board.stop_execution()
print("Stop Execution:", execution)

program_files = board.load_program_file(0)
print("Program Files:", program_files)

load = board.load_correction_file(0,1,2)
print("Correction File Load:", load)

table = board.select_cor_table(1,0)
print("Correction Table Selection:", table)
```

Alterar de modo:

```python
laserbox.write(f"SOURce2:AM:EXTernal DIGital")
laserbox.write(f"SOURce3:AM:EXTernal DIGital")
laserbox.write(f"SOURce4:AM:EXTernal DIGital")
laserbox.write(f"SOURce5:AM:EXTernal DIGital")
```

Como saber em qual modo está o laser?

```python
laserbox.query(f"SOURce2:AM:SOURce?")
```

```python
laser = 5
power = 1/1000
laserbox.query(f"SOURce{laser}:AM:STATe ON")
laserbox.query(f"SOURce{laser}:POWer:LEVel:IMMediate:AMPLitude %.5f" % power)
```
